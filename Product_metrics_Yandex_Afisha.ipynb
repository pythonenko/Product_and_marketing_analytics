## Рассчет метрик

### Продуктовые метрики

#### DAU, WAU и MAU

```python
monthly_stat = visits.groupby(
    pd.Grouper(key='start_ts', freq='1M')).agg({'uid':['count', 'nunique']}).reset_index()
monthly_stat['start_ts'] = monthly_stat['start_ts'].dt.strftime('%Y-%m')
monthly_stat.columns = ['date','n_sessions', 'mau']

weekly_stat = visits.groupby(
    pd.Grouper(key='start_ts', freq='w')).agg({'uid':['count', 'nunique']}).reset_index()
weekly_stat.columns = ['date','n_sessions', 'wau']

daily_stat = visits.groupby(
    pd.Grouper(key='start_ts', freq='d')).agg({'uid':['count', 'nunique']}).reset_index()
daily_stat.columns = ['date','n_sessions', 'dau']

mau_total = monthly_stat['mau'].mean()
wau_total = weekly_stat['wau'].mean()
dau_total = daily_stat['dau'].mean()

print('Среднее значение MAU:', int(mau_total)) 
print('Среднее значение WAU:', int(wau_total)) 
print('Среднее значение DAU:', int(dau_total))

sticky_wau = dau_total/wau_total
sticky_mau = dau_total/mau_total

print('Недельная вовлеченность аудитории: {:.2%}'.format(sticky_wau))
print('Месячная вовлеченность аудитории: {:.2%}'.format(sticky_mau))
```

```python
fig = go.Figure([go.Scatter(x=monthly_stat['date'], 
                            y=monthly_stat['mau'])])
fig.update_layout(title_text='MAU', title_font_size=20,
                  xaxis_title='month',
                  yaxis_title="MAU")
fig.update_xaxes(dtick="M1", tickformat="%b\n%Y")
fig.show(width=1000, height=500, renderer='svg')
```

```python
fig = go.Figure([go.Scatter(x=weekly_stat['date'], 
                            y=weekly_stat['wau'])])
fig.update_layout(title_text='WAU', title_font_size=20,
                  xaxis_title='week',
                  yaxis_title="WAU")
fig.update_xaxes(dtick=86400000.0 * 14, tickformat="%W\n%b", tickangle = 0)
fig.show(width=1000, height=500, renderer='svg')
```

```python
fig = go.Figure([go.Scatter(x=daily_stat['date'], 
                            y=daily_stat['dau'])])
fig.update_layout(title_text='DAU', title_font_size=20,
                  xaxis_title='day',
                  yaxis_title="DAU")
fig.update_xaxes(dtick=86400000.0 * 7, tickformat="%d-%b", tickangle = 270)
fig.show(width=1000, height=500, renderer='svg')
```

**Вывод.** Анализ пользовательской активности показал, что в среднем к сервису Яндекс.Афиша обращаются 23228 уникальных пользователей в месяц, 5716 уникальных пользователей в неделю и 905 уникальных пользователей в день.

Пиковая пользовательская активность наблюдалась в ноябре, в 47 неделю, рекордная посещаемость зафиксирована 24 ноября (3319). 
Минимальная активность наблюдалась в августе, в 31 неделю. Антирекорд по дневной посещаемости установлен 31 марта и 1 апреля - вероятно, у сервиса был технический сбой в эти дни.

Значения показателей sticky factor говорят о низкой вовлеченности аудитории: всего 15.8% для недельной аудитории и 3.9% - для месячной. Эти показатели говорят о том, что каждый день аудитория сервиса почти полностью обновляется (если сравнивать с 1/30 и 1/7 соответственно).

#### Время на сайте

Исследуем продолжительность типичной пользовательской сессии за весь период. Чтобы выбрать подходящую среднюю меру, построим график распределения.

```python
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds

visits.plot(y = 'session_duration_sec', kind = 'hist', bins = 100, color = 'peru',
            legend=None, grid=True, figsize = (10,4), range = (0,10000))
plt.xlabel('Продолжительность сессии, секунд')
plt.ylabel('Частота')
plt.title('Продолжительность сессий')
plt.show()

display(visits['session_duration_sec'].describe())
```

Распределение продолжительности пользовательских сессий сильно отличается от нормального: из-за присутствия выбросов наблюдается явное смещение вправо, а среднее значение больше медианы почти в два раза. Кроме того, в базе есть записи с нулевой продолжительностью сессий. Поэтому в качестве средней меры ASL вместо среднего или медианы будем использовать моду.

```python
asl = (visits.query('session_duration_sec > 0')['session_duration_sec']).mode()
print('Средняя продолжительность сессии составляет {} секунд.'.format(asl[0]))
```

**Вывод.** В среднем сессия пользователя длится 1 минуту. Чтобы это выяснить, изучим метрики электронной коммерции, в частности, мтерику времени от первого визита до покупки.

#### Retention Rate

Применяя когортный анализ, рассчитаем Retention Rate и изобразим изменения метрики во времени на графике.

```python
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')

first_activity_month = visits.groupby(['uid'])['visit_month'].min()
first_activity_month.name = 'first_visit_month'

visits = visits.join(first_activity_month, on='uid')

visits['cohort_lifetime'] = (visits['visit_month'] - visits['first_visit_month'])/ np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')

cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index() 

initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
cohorts = cohorts.merge(initial_users_count,on='first_visit_month') 
cohorts['retention'] = (cohorts['uid']/cohorts['cohort_users']).round(4)

retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)
retention_pivot.index = retention_pivot.index.strftime('%Y-%m')

sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', 
            vmin=0, vmax=0.1)
plt.show()
```

Найдем средний Retention Rate на второй месяц «жизни» когорт.

```python
print("Retention Rate на второй месяц «жизни» когорт: {0:.2f}%".format(retention_pivot[1].mean()*100))
```

**Вывод.** 
Средний Retention Rate второго месяца составил 6.52%. Наблюдается заметная негативная динамика RR второго месяца. Так, когорты, привлеченные с августа по ноябрь, удерживали на 2 месяц не менее 7.7%, в то время как RR декабрьской и последующих когорт не превышал 6% и постепенно снижался.

### Метрики электронной коммерции

#### Среднее время с первого посещения сайта до совершения покупки

Исследуем, сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки.

```python
first_visit_date = visits.groupby(['uid'])['start_ts'].min()
first_visit_date.name = 'first_visit_datetime'

visits = visits.join(first_visit_date, on='uid')

first_purchase = orders.groupby('uid')['buy_ts'].min().to_frame()
first_purchase.columns = ['first_purchase_datetime']
first_purchase = pd.merge(first_purchase, first_visit_date, left_index=True, right_index=True, how='left')
first_purchase['time_to_buy']= \
(first_purchase['first_purchase_datetime']-first_purchase['first_visit_datetime']).dt.seconds//60

display(first_purchase['time_to_buy'].describe().round())

first_purchase['time_to_buy'].plot(kind = 'hist', bins = 50, color = 'peru',
            legend=None, grid=True, figsize = (10,4), range = (0,250))
plt.xlabel('Время, минут')
plt.ylabel('Частота')
plt.title('Время до первой покупки')
plt.show()
```

```python
time_to_buy = first_purchase['time_to_buy'].round(3).mode()
print('Модальное значание времени до покупки составляет {} минуту.'.format(time_to_buy[0]))
```

**Вывод.**

#### Среднее количество покупок на одного покупателя
Рассчитаем среднее количество покупателей и совершенных ими покупок для когорт в разрезе месяцев.

```python
first_purchase_dates = orders.groupby('uid')['buy_ts'].min()
first_purchase_dates.name = 'first_purchase_datetime'
orders = orders.join(first_purchase_dates, on='uid')

orders['purchase_month'] = orders['buy_ts'].astype('datetime64[M]')
orders['first_purchase_month'] = orders['first_purchase_datetime'].astype('datetime64[M]')

purchases_grouped_by_cohorts = orders.groupby(
    ['first_purchase_month', 'purchase_month']
)['uid'].agg(['count', 'nunique']).reset_index()

purchases_grouped_by_cohorts['cohort_lifetime'] = (
    purchases_grouped_by_cohorts['purchase_month']
    - purchases_grouped_by_cohorts['first_purchase_month'])

purchases_grouped_by_cohorts['cohort_lifetime'] = purchases_grouped_by_cohorts[
    'cohort_lifetime'] / np.timedelta64(1, 'M')
purchases_grouped_by_cohorts['cohort_lifetime'] = (
    purchases_grouped_by_cohorts['cohort_lifetime'].round().astype('int'))
```

Применяя когортный анализ, рассчитаем количество покупок за 6 месяцев.

```python
six_month_cohorts = purchases_grouped_by_cohorts[(
    purchases_grouped_by_cohorts['first_purchase_month'] < '2018-01-01')& (
    purchases_grouped_by_cohorts['cohort_lifetime'] < 6)]

lifetime_pivot = six_month_cohorts.pivot_table(
    index='first_purchase_month', columns='cohort_lifetime', values='count',aggfunc='sum')

lifetime_pivot.index = lifetime_pivot.index.strftime('%y-%m')

sns.set(style='white')
plt.figure(figsize=(10, 7))
plt.title('Cohorts: Orders')
sns.heatmap(lifetime_pivot, annot=True, fmt='1', vmax=500, 
            vmin=100, center= 300, cmap= 'coolwarm')
plt.show()
```

С сентября 2017 года число покупок в первый, второй и третий месяцы росло. У большинства когорт число покупок заметно снижается на 5 и 6 месяцы.

Рассчитаем среднее количество покупок на одного пользователя за 6 месяцев.

```python
lifetime_pivot['sum'] = lifetime_pivot.sum(axis=1)

n_users = six_month_cohorts[six_month_cohorts['cohort_lifetime'] == 0].pivot_table(
    index='first_purchase_month', values=['nunique'], aggfunc='sum')
n_users.index = n_users.index.strftime('%y-%m')

lifetime_pivot = lifetime_pivot.merge(n_users,left_index=True, right_index=True)
lifetime_pivot['purchases_per_users'] = (lifetime_pivot['sum']/lifetime_pivot['nunique']).round(2)
```

```python
display(lifetime_pivot[['sum', 'nunique','purchases_per_users']])
print('Покупок на пользователя в среднем за 6 месяцев:', lifetime_pivot['purchases_per_users'].mean().round(2))
```

#### LTV

Рассчитаем средний LTV по когортам за 6 месяцев.

```python
cohort_sizes = (
    orders.groupby('first_purchase_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_purchase_month', 'n_buyers']

margin_rate = 1

cohorts = (
    orders.groupby(['first_purchase_month', 'purchase_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
report = pd.merge(cohort_sizes, cohorts, on='first_purchase_month')
report['gp'] =  report['revenue'] * margin_rate
report['age'] = (report['purchase_month'] - report['first_purchase_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int') 
report['ltv'] = report['gp'] / report['n_buyers']

six_month_report = report[(report['first_purchase_month'] < '2018-01-01')& (report['age'] < 6)]

result = six_month_report.pivot_table(
    index='first_purchase_month', columns='age', values='ltv', aggfunc='mean'
).round(2)

result = result.cumsum(axis=1).round(2)

result.index = result.index.strftime('%Y-%m')
```

Отразим на графике изменения метрики LTV во времени по когортам.

```python
sns.set(style='white')
plt.figure(figsize=(9, 7))
plt.title('Cohorts: LTV')
sns.heatmap(result, annot=True, fmt='.2g', 
            vmin=4, vmax=12, cmap= 'coolwarm')
plt.show()
```

С точки зрения накопленного за 6 месяцев LTV сентябрьская, и в меньшей степени июньская, когорты являются наиболее ценными для бизнеса. В ноябре и октябре были привлечены пользователи с наименьшим потенциалом LTV.
