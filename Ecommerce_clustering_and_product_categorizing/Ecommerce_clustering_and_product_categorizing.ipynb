# Описание проекта

**Цель проекта:**

Используя информацию о транзакциях интернет-магазина товаров для дома и быта «Пока все ещё тут», рассчитаем основные метрики для каждого пользователя и проведем сегментацию покупателей с применением кластерного анализа на основе истории их покупок. Для оценки значимости различий ключевых метрик между сегментами также будет проведена серия статистических тестов с проверкой гипотез по критерию Манна-Уитни. Чтобы упростить процедуру кластеризации и интерпретируемость результатов, товары будут предварительно сгруппированы в несколько подкатегорий, соответственно, покупки пользователей также можно будет рассматривать укрупненно и объединять клиентов с одинаковыми предпочтениями в сегменты.

Кластеризация покупателей по профилю потребления позволит выделить однородные по продуктовым предпочтениям группы для дальнейшей разработки персонализированных предложений таргетированной рекламы. Также эта информация может быть использована для создания виджетов с персональными рекомендациями на сайте или в приложении интернет-магазина или отправки персональных предложений через push-уведомления или email-рассылки. Кроме того, кластерный анализ позволит идентифицировать целевые группы клиентов, чтобы в дальнейшем инвестировать в повышение их лояльности и таргетироваться на похожих пользователей при привлечении новых клиентов. Типичный портрет непривлекательных клиентов, в свою очередь, поможет бизнесу меньше тратить на их привлечение и удержание.

**Описание данных:**

При анализе данных будет использован датасет *ecommerce_dataset.csv*, который содержит информацию о транзакциях интернет-магазина, где:
- `date` — дата заказа;
- `customer_id` — идентификатор покупателя;
- `order_id` — идентификатор заказа;
- `product` — наименование товара;
- `quantity` — количество товара в заказе;
- `price` — цена товара.

Приведем значения столбца product к нижнему регистру и рассчитаем число уникальных SKU.

Для упрощения дальнейшего анализа проведем категоризацию продуктов интернет-магазина. Для этого создадим отдельный словарь с уникальными названиями продуктов. Также выделим 1е и 2е слово из названия в отдельные столбцы, так как тип большинства продуктов можно понять по первому слову (а уникальных значений гораздо меньше, чем названий продуктов), а для остальных можно использовать второе слово для уточниния, после чего лемматизируем все значения. Ожидается, что количество уникальных первых слов после лемматизации будет существенно меньше исходного перечня SKU, что упростит процесс группировки.

```python
m = Mystem()

product_dict = data['product'].drop_duplicates().reset_index()
product_dict['first_word'] = product_dict['product'].str.findall(r'\w+').str[0]
product_dict['second_word'] = product_dict['product'].str.findall(r'\w+').str[1]
product_dict['lemma'] = product_dict['product'].apply(m.lemmatize)
product_dict['first_word'] = product_dict['first_word'].apply(m.lemmatize).str[0]
product_dict['second_word'] = product_dict['second_word'].apply(m.lemmatize).str[0]

print("Уникальных значений 1го слова в названиях: ", product_dict['first_word'].nunique())
display(product_dict.head())
```

##  Сегментация покупателей на основе истории их покупок

### Расчет поюзерных метрик

Рассчитаем число заказов и покупок, среднее количество товаров, сумму покупок и число уникальных категорий для каждого покупателя.

```python
users_stat = data.groupby('customer_id').agg({'order_id': 'nunique',
                                              'product': 'count',
                                              'quantity': 'mean',
                                              'total': 'sum',
                                              'category': 'nunique'
                                              }).reset_index()

users_stat.columns = ['customer_id', 'orders', 
                      'purchases','avg_quantity', 'total_sum', 'n_categories']
```

Дополнительно рассчитаем метрику среднего чека.

```python
users_stat['avg_check'] = users_stat['total_sum']/users_stat['orders']

display(users_stat.describe().T.round(2))
```

В среднем каждый пользователь совершил по 1 заказу и 2 покупки в одной товарной категории, купив по 2 единицы каждого SKU на общую сумму 1323 рубля, при этом средний чек составил 1139 рублей.
Наличие выбивающихся значений числа заказов и покупок могут повлиять на результат кластерного анализа: выброс сформирует отдельный сегмент, а остальные наблюдения объединятся из-за относительной близости значений. Чтобы этого избежать, удалим выбивающиеся значения выше 99й квантили.

```python
orders_cut = np.percentile(users_stat['orders'], 99)
purchases_cut = np.percentile(users_stat['purchases'], 99)

users_stat = users_stat.loc[
    (users_stat['orders'] <= orders_cut)& 
    (users_stat['purchases'] <= purchases_cut)]
```

Создадим еще одну таблицу с разбиением на категории, где для каждого покупателя будет указано 1, если он хотя бы раз покупал товар из соответствующей категории, и 0, если нет. Это упростит дальнейшую интерпретацию центров кластеров как долю представителей кластера, купиших продукт хотя бы 1 раз.

```python
users_categ = data.pivot_table(
    index='customer_id', 
    columns = 'category', 
    values=['order_id'], 
    aggfunc=({'order_id': 'nunique'}), fill_value=0
   )

users_categ = users_categ.replace(0, np.nan).notnull().astype(int).reset_index()

users_categ.columns = ['customer_id' , 'interior', 'flowers', 'dishes_and_kitchen',
                       'garden', 'household_chemicals', 'renovation', 'textiles',
                       'сlothing_and_footwear_care', 'household_goods']
```

Объединим 2 таблицы по id пользователей.

```python
users_stat = users_stat.merge(users_categ, how = 'left', on='customer_id')
```

### Сегментация клиентов на основе кластерного анализа

Поскольку анализ будет строиться на небольшом числе наблюдений, для лучшей кластеризации следует избавиться от лишних переменных. Для этого изучим матрицу корреляций: 

```python
corr = users_stat.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(14, 10))
    ax = sns.heatmap(corr, annot=True, fmt='.2f', vmin=-0.4, vmax=1, cmap='inferno_r', 
                     center=-0.1, mask=mask, square=True)
plt.show()
```

Переменные total_sum и avg_check сильно коррелированы, поэтому для кластеризации можно оставить одну из них. С позиций юнит-экономики правильнее использовать комбинацию среднего чека и числа заказов вместо общей суммы.

Удалим лишние переменные:

```python
X = users_stat.drop(['customer_id', 'total_sum'], axis=1)
X.info()
```

Стандартизируем данные для проведения кластеризации:

```python
sc = StandardScaler()
x_sc = sc.fit_transform(X)
```

На основе полученного графика иерархической кластеризации можно предварительно выделить 11 кластеров, поскольку после этого шага агломерации происходит заметный рост расстояний между кластерами. Кроме того, при 10 и менее кластерах относительные размеры групп будут сильно отличаться.

Однако такое количество кластеров будет сложно интерпретировать, поэтому для определения оптимальнго числа кластеров также применим метод "согнутого колена", подразумевающий многократное циклическое исполнение алгоритма кластеризации методом k-средних с увеличением количества выбираемых кластеров, а также последующим откладыванием на графике значения метрики SSE (суммы квадратов ошибок).

```python
K = range(1,15)
sse = []
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=0)
    kmeanModel.fit(x_sc)
    sse.append(kmeanModel.inertia_)

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(K, sse, 'bx-', color = '#c43c4e')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Как такового перелома в графике с ростом числа кластров не происходит, однако снижение метрики SSE замедляется с переходом от 10 к 11 кластерам. Поэтому можно считать 10 оптимальным числом кластеров.

Дополнительно применим метод Силуэта:

```python
k_clusters = []
sil_coeffecients = []

for n_cluster in range(2,15):
    kmeans = KMeans(n_clusters = n_cluster, random_state=0).fit(x_sc)
    label = kmeans.labels_
    sil_coeff = silhouette_score(x_sc, label)
    print("For n_clusters={}, Silhouette Coefficient = {}".format(n_cluster, sil_coeff))
    sil_coeffecients.append(sil_coeff)
    k_clusters.append(n_cluster)
    
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(k_clusters, sil_coeffecients, 'bx-', color = '#87216b')
plt.title('Silhouette Method showing the optimal k')
plt.ylabel('Silhouette Coefficient')
plt.xlabel('k')
plt.show()

sil_best = max(sil_coeffecients)
k_best_index = sil_coeffecients.index(sil_best)
k_best = k_clusters[k_best_index]
print("Оптимальное число кластеров:", k_best)
```

Оптимальное количество кластеров по методу Силуэта равняется 13 и продолжает расти, однако можно заметить, что рост метрики замедляется на 7 кластерах. Остановимся на этом количестве, как оптимальном с точки зрения интерпретируемости результатов и качества подгонки модели.
Обучим модель кластеризации на основе алгоритма K-Means с числом кластеров, равным 7, и спрогнозируем кластеры для наблюдений:

##  Проверка статистических гипотез

Для каждой пары кластеров проверим гипотезу "Распределение [метрики] кластеров различается" при помощи непараметрического теста Манна-Уитни:
    
* H0 - нет статистически значимых различий между распределениями метрики;
    
* H1 - есть статистически значимые различия между распределениями метрики.

Проведенный исследовательский анализ данных говорит о сильной вариации в данных, поэтому минимизируем риск обнаружения ложных межгрупповых отличий и установим пороговое значение alpha, равным 0.05, то есть будем отвергать нулевую гипотезу, если вероятность обнаружить межргупповые различия при их отсутствии в генеральной совокупности меньше 5%.

Создадим функцию для расчета статистической значимости различий между группами, но перед этим рассчитаем количество сравнений для поправки Бонферрони для множественных сравнений:

```python
n_cl = np.sort(X['cluster_km'].unique())
n_mw_compared = len(list(combinations(n_cl, 2)))*5
print('Число проверок гипотез: ', n_mw_compared)
```

```python
def mannwhitneyu_test (alpha, variable, first_cluster, second_cluster):
    results = st.mannwhitneyu(
        X[X['cluster_km']== first_cluster][variable], 
        X[X['cluster_km']== second_cluster][variable])
    pvalue = results.pvalue.round(4)
    bonferroni_alpha = alpha/n_mw_compared            
    if results.pvalue < bonferroni_alpha:
        result = "Отвергаем нулевую гипотезу"
    else:
        result = "Не получилось отвергнуть нулевую гипотезу"
    return[pvalue, result]
```

### Отличия кластеров по метрикам электронной коммерции

Создадим функцию для вывода результатов в сводную таблицу, фильтрующую только значимые отличия между кластерами:

```python
def mw_table (variable):
    cl = np.sort(X['cluster_km'].unique())
    comb = pd.DataFrame(combinations(cl, 2))
    comb.columns = ['cl_0', 'cl_1']
    split = pd.DataFrame(
        comb.apply(lambda x: mannwhitneyu_test(0.05, variable, x.cl_0, x.cl_1), axis=1).tolist(),
        columns=['pvalue', 'result'])
    comb = pd.concat([comb, split], axis=1)
    display(comb.query('result == "Отвергаем нулевую гипотезу"'))
```

### Отличия кластеров по товарным предпочтениям

Проверим статистическую достоверность отличий в долях покупатетей каждой категории товаров для пар кластеров при помощи z-теста для проверки равенства долей:
* H0 - нет статистически значимых различий между долями для представителей первого и второго кластра;
* H1 - есть статистически значимые различия между долями для представителей первого и второго кластра.

Создадим функцию для проверки подобных гипотез с поправкой на множественные сравнения, фильтрующую пары кластеров со значимыми отличиями:

```python
n_z_compared = len(list(combinations(n_cl, 2)))*9
print('Число проверок гипотез: ', n_z_compared)
```

```python
def z_test (alpha, variable, first_group, second_group):
    
    total_first_group = len(X.loc[X['cluster_km'] == first_group])
    total_second_group = len(X.loc[X['cluster_km'] == second_group])
    
    converted_first_group = X.loc[X['cluster_km'] == first_group][variable].sum()
    converted_second_group = X.loc[X['cluster_km'] == second_group][variable].sum()
    
    if (converted_first_group !=0) and (converted_second_group !=0):
        p1 = X.loc[X['cluster_km'] == first_group][variable].mean()
        p2 = X.loc[X['cluster_km'] == second_group][variable].mean()
        p_combined = (converted_first_group + converted_second_group) / (total_first_group + total_second_group)
        difference = p1 - p2

        z_value = difference / math.sqrt(
            p_combined * (1 - p_combined) * (1/total_first_group + 1/total_second_group))

        distr = st.norm(0, 1)  
        pvalue = ((1 - distr.cdf(abs(z_value))) * 2).round(4)

        bonferroni_alpha = alpha/n_z_compared

        if pvalue < bonferroni_alpha:
            result = 'Отвергаем нулевую гипотезу: между долями есть значимая разница'
        else:
            result = 'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    else: 
        pvalue = "NaN"
        result = 'NaN'
        
    return[pvalue, result]
```

Создадим функцию для вывода результатов в сводную таблицу:

```python
def z_table (variable):
    cl = np.sort(X['cluster_km'].unique())
    comb = pd.DataFrame(combinations(cl, 2))
    comb.columns = ['cl_0', 'cl_1']
    split = pd.DataFrame(
        comb.apply(lambda x: z_test(0.05, variable, x.cl_0, x.cl_1), axis=1).tolist(),
        columns=['pvalue', 'result'])
    comb = pd.concat([comb, split], axis=1)
    display(comb.query('result == "Отвергаем нулевую гипотезу: между долями есть значимая разница"'))
```

###  Рекомендации

На основе результатов проведенной сегментации клиентов интернет-магазина товаров для дома и быта «Пока все ещё тут» можно предложить команде по управлению пользовательским опытом следующие рекомендации:

**В части таргетированной рекламы** рекомендуется персонализировать предложения в рамках категорий товаров следующим образом: 
* 1 кластер: хозяйственные товары
* 2 кластер: комнатные цветы и текстиль
* 3 кластер: товары для строительства и ремонта, текстиль, товары для ухода за одеждой и обувью
* 4 кластер: посуда и товары для кухни 
* 5 кластер: текстиль и садовые растения
* 6 кластер: средства гигиены, бытовая химия и товары для уборки
* 7 кластер: товары для интерьера и хранения в доме 

Также можно использовать полученные группы покупателей, однородные по поведению и продуктовым предпочтениям, для создания виджетов с персональными рекомендациями на сайте или в приложении интернет-магазина или для отправки персональных предложений через push-уведомления или email-рассылки. Такие кампании могут быть посвящены как самим продуктам в ассортименте магазина - новинки, наиболее популярные SKU среди представителей кластера, так и проводимым акциям и персональным скидкам/промокодам.

**В части идентификации целевых групп клиентов** рекомендуется в первую очередь ориентироваться на:
* покупателей из 5 кластера, совершающих больше всего транзакций
* покупателей из 1 кластера, имеющих максимальный средний чек
* покупателей из 2 и 3 кластеров, совершающих больше всего заказов

Для указанных целевых групп клиентов следует разработать меры по повышению лояльности, такие как персонализированные скидки и промокоды, начисление баллов за обратную связь о качестве товаров и сервиса, закрытый доступ к распродажам. Кроме того, при использовании дополнительных характеристик этих клиентов (например, указанных при регистрации или из счетчиков сайта) можно также сформировать более полный портрет целевых групп клиентов, что поможет таргетироваться на похожих пользователей в рекламных кампаниях.

Что касается покупателей с низкой привлекательностью для бизнеса, рекомендуется также разработать механизмы по стимулированию целевых действий для их представителей, соответствующих метрикам электронной коммерции: повторные покупки (все кластеры кроме 3), средний чек (все кластеры кроме 1) и число товаров в одном заказе (все кластеры кроме 2 и 5). Например, для стимулирования роста среднего чека можно установить порог минимальной суммы заказа для бесплатной доставки на уровне среднего чека 1 сегмента (2000 рублей), а для стимулирования повторных покупок - давать сгорающий кэшбэк на следующие заказы. Вместе с тем, при обогащении типичного портрета непривлекательных клиентов информацией из счетчиков и CRM-системы бизнесу следует меньше инвестировать в привлечение клиентов с похожими характеристиками.

**В части развития инструментов персонализации** также рекомендуется не только предлагать клиентам из одного кластера продукты, купленные другими его представителями, но и персонализировать рекомендации на микроуровне. Так, при накоплении статистики повторных покупок можно предлагать покупателям одинаковых товаров другие продукты из истории транзакций друг друга.
