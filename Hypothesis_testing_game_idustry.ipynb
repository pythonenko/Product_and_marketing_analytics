# Анализ факторов успешности игр #

На основе данных о глобальных продажах игр до 2016 года проанализируем, какие факторы определяют коммерческую упешность игр, какова роль возрастного рейтинга ESRB и оценок пользователей и экспертов, какие платформы и жанры являются наиболее перпективными, а также изучим региональную специфику игровых рынков Северной Америки, Европы и Японии. Это позволит интернет-магазину «Стримчик» принимать обоснованные решения о выборе продуктов и рекламных кампаниях в 2017 году.

```python
import pandas as pd
import numpy as np
from scipy import stats as st
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import math
from IPython.display import display
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import os
```

## Изучение данных

Откроем файл с данными и изучим общую информацию:

```python
server_path = '/datasets/games.csv'
filename = 'games.csv'
    
if os.path.exists(server_path):
    data = pd.read_csv(server_path)
else:
    data = pd.read_csv(filename)
    
data.info()
display(data.head())
```

В таблице содержится 16715 записей и 11 переменных. Обнаружены пропуски в переменных: Name, Year_of_Release, Genre, Critic_Score, User_Score, Rating. Названия переменных для удобства следует перевести в нижний регистр. Кроме того, типы переменных Year_of_Release и user_score следует перевести в целочисленный формат.

### Обработка пропусков ###

Изучим наблюдения с пропусками в различных переменных.

```python
display(data[data['name'].isnull()])
```

В данных обнаружено 2 игры без названия. Поскольку дальнейший анализ будет проводиться на уровне платформ и жанров, а не отдельных игр, можно сохранить строки, перекодировав пустые значения на Unknown.

```python
display(data[data['year_of_release'].isnull()].head())
```

Год релиза будет использоваться в качестве критерия фильтрации актуальных данных, значит, анализировать данные без года невозможно. 269 пропусков в переменной year_of_release можно частично восстановить по переменной name (например, Madden NFL 2004) или на основе открытых данных. Однако проще присвоить таким данным маркер, поскольку таких записей всего 1,6%.

```python
display(data[data['critic_score'].isnull()].head())
display(data[data['user_score'].isnull()].head())
display(data[data['rating'].isnull()].head())
```

Пропуски в переменных critic_score, user_score, rating могут быть связаны с непопулярностью игр или с неразвитостью систем рейтингования в годы выхода игр, например, 1985 год. Такие наблюдения тоже представляют ценность для анализа (когда рейтинг не учитывается), поэтому также заменим пропуски на маркер.

Переменная user_score распознана как категориальная, изучим причину аномалии:

```python
print(data['user_score'].unique())
```

Обнаружено значение tbd, то есть To Be Determined (будет определено), что фактически означает пропуски в данных. Проведем замену в 2 этапа: сначала заменим на классичекое NaN, а после заменим на маркер параллельно с обработкой пропусков в других переменных.

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan, regex=True)
```

Добавим маркеры на место всех пропусков: для категориальных переменных будем использовать Unknown, для количественных - 999.

```python
data = data.fillna({'name': 'Unknown', 'critic_score': 999, 'user_score': 999, 'rating': 'Unknown', 'year_of_release': 999})
data.info()
```

Избавились от пропусков во всех переменных.

### Выявление перспективных платформ ###

На основе установленного характерного жизненного цикла платформ отфильтруем данные за актуальный период (4 полных года), то есть с 2012 года включительно.

```python
actual_data = data.query('year_of_release >= 2012')
```

Чтобы определить потенциально прибыльные платформы, проанализируем динамику продаж за актуальный период.

```python
actual_data_pivot = actual_data.pivot_table(
    index = 'year_of_release', columns='platform', 
    values = 'total_sales', aggfunc = 'sum', margins=True).fillna(value=0)
display(actual_data_pivot)

platforms = actual_data['platform'].unique()

size = math.ceil(math.sqrt(len(platforms)))

all_sales = actual_data.groupby(['platform','year_of_release']).agg({'total_sales' : 'sum'}).reset_index()

fig = make_subplots(
    rows=size, cols=size,
    subplot_titles=(platforms))

row = 0
col = 0

for i,p in enumerate(platforms):        
    
    if i % size == 0:
        col = 1
        row += 1
    else:
        col += 1
    data_ = all_sales[all_sales['platform'] == p]
    fig.add_trace(go.Scatter(x = data_['year_of_release'], y = data_['total_sales']),
              row=row, col=col)
    
fig.update_xaxes(range=[2012, 2015])
fig.update_layout(height=1700, width=1500,
                  title_text="Распределение продаж по годам")

fig.show(width=1000, height=1000, renderer='svg')
```

По суммарной выручке за 4 года лидерами являются PS4, PS3, X360, 3DS и XOne. PS4 опережает другие платформы с 2014 года. Учитывая неполные данные за 2016 год, XOne и PS4 являются единственными растущими платформами. Остальные платформы теряют рынок, DS и PSP не приносят продаж с 2013 и 2015 года соответственно. Также по таблице видно, что PS4 опередила PS3 по доходам от игр, выпущенных с 2014 года, а XOne опередила X360.

### Влияние отзывов на продажи ###

Для начала подготовим данные. Исключим DS и Wii по причине малого числа наблюдений, а также игры с маркерами в переменных critic_score user_score. И удалим выбросы.

```python
score_data = actual_data.query('critic_score != 999 and user_score != 999 and platform != ["DS", "Wii", "PSP"]')

def cut_outlier(s):
    lower_limit = s.mean() - (s.std() * 3)
    upper_limit = s.mean() + (s.std() * 3)
    return ~s.between(lower_limit, upper_limit)

score_data = score_data[~score_data.groupby('platform')['total_sales'].apply(cut_outlier)]
```

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

```python
PS4 = score_data.query('platform == "PS4"')[['total_sales','critic_score','user_score']]
pd.plotting.scatter_matrix(PS4, figsize=(9, 9))
display(PS4.corr())
```

Для игр, выпущенных на PS4, характерна слабая положительная связь продаж с отзывами критиков, при этом влияние пользовательских оценок отсутствует. Вместе с тем, отзывы экспертов и пользователей сильно коррелированы.

## Проверка гипотез

### Сравнение пользовательских рейтингов платформ Xbox One и PC ###

```python
df = actual_data.query('user_score != 999 and platform == ["XOne", "PC"]')
display (df.groupby('platform')['user_score'].describe().round(2))
```

Средние выборочные (а также медианные) значения по двум платформам отличаются (на 5,5% выше среди игр на PC в сравнении с XOne), однако какова верояность получить такие различия случайно?

Проверим гипотезу "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" (это нулевая гипотеза). Альтернативная гипотеза: различия между пользовательскими рейтингами есть.

Поскольку средние значения оценок пользователей по группам очень близки и стандратное отклонение внутри обеих групп говорит о сильной вариации в данных, минимизируем риск обнаружения ложных межгрупповых отличий и установим пороговое значение alpha, равным 0.01, то есть будем отвергать нулевую гипотезу, если вероятность обнаружить межгрупповые различия при их фактическом отсутствии меньше 1%.

```python
alpha = 0.01
results = st.ttest_ind(
    df.query("platform == 'XOne'")['user_score'], 
    df.query("platform == 'PC'")['user_score'],
equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод.**  Полученное значение p-value говорит о том, что хотя средние оценки игр на PC выше, с вероятностью 3.5% такое или большее различие можно получить случайно. Это невысокая вероятность (на уровне alpha 0,05 мы бы отвергли нулевую гипотезу), но делать вывод о значимом различии между средними все же не будем, отвергаем нулевую гипотезу, значимых отличий пользовательских оценок между PC и XOne нет.

### Сравнение пользовательских рейтингов платформ Xbox One и PC ###

```python
df = actual_data.query('user_score != 999 and genre == ["Action", "Sports"]')
display (df.groupby('genre')['user_score'].describe().round(2))
```

Различие средних пользовательских оценок между жанрами Action и Sports более выражено: оценки игр в жанре Action на 6,3% выше.

Проверим гипотезу "Средние пользовательские рейтинги жанров Action и Sports разные" (альтернативная гипотеза). Нулевая гипотеза: различий в оценках между жанрами нет.

Также будем использовать строгий критерий для проверки гипотезы (alpha=0.01), поскольку средние значения по группам близки и, учитывая сильную вариацию в данных, велик риск обнаружить ложноположительные отличия.

```python
results = st.ttest_ind(
    df.query("genre == 'Action'")['user_score'], 
    df.query("genre == 'Sports'")['user_score'],
equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод.**
Полученное значение p-value говорит о том, что выборочные средние достаточно далеки друг от друга, чтобы отбросить нулевую гипотезу об их равестве (вероятность получить такое различие случайно крайне мала). Значит, средняя пользовательская оценка игр в жанре Action статистически значимо отличается от средних оценок жанра Sports.
